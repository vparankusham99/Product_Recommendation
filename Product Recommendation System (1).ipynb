# Train-Test Splitting

```python
#Splitting the Data
from sklearn.model_selection import train_test_split

#Split the data randomnly into train and test datasets into 70:30 ratio
train_data, test_data = train_test_split(df2, test_size = 0.3, random_state=0)
train_data.head()
```

```python
print('Shape of training data: ',train_data.shape)
print('Shape of testing data: ',test_data.shape)
```

```python
#Count of user_id for each unique product as recommendation score 

train_data_grouped = train_data.groupby('PRODUCTID').agg({'USERID': 'count'}).reset_index()
train_data_grouped.rename(columns = {'USERID': 'score'},inplace=True)
train_data_grouped.head(40)
```

```python
#Sort the products on recommendation score 
train_data_sort = train_data_grouped.sort_values(['score', 'PRODUCTID'], ascending = [0,1]) 
      
#Generate a recommendation rank based upon score 
train_data_sort['rank'] = train_data_sort['score'].rank(ascending=0, method='first') 
          
#Get the top 5 recommendations 
popularity_recommendations = train_data_sort.head(5) 
popularity_recommendations 
```

```python
# Use popularity based recommender model to make predictions
def recommend(user_id):     
    user_recommendations = popularity_recommendations 
          
    #Add user_id column for which the recommendations are being generated 
    user_recommendations['USERID'] = user_id 
      
    #Bring user_id column to the front 
    cols = user_recommendations.columns.tolist() 
    cols = cols[-1:] + cols[:-1] 
    user_recommendations = user_recommendations[cols] 
          
    return user_recommendations 
```

```python
import warnings
warnings.filterwarnings('ignore')
find_recom = [10,100,150]   # This list is user choice.
for i in find_recom:
    print("The list of recommendations for the UserID: %d\n" %(i))
    print(recommend(i))    
    print("\n") 
```

# COLLABORATIVE FILTERING RECOMMENDER MODEL

```python
Amazon_products = pd.concat([train_data, test_data]).reset_index()
Amazon_products.head()
```

```python
#Taking the first 10000 entries as sample Dataframe
Amazon_ratings = df2.head(10000)
```

```python
#Creating a pivot table
ratings_matrix = Amazon_ratings.pivot_table(values='RATINGS', index='USERID', columns=['PRODUCTID','LINKS'], fill_value=0)
ratings_matrix.head()
```

```python
#Shape of the matric created.
ratings_matrix.shape
```

```python
#Transposing the matrix 

X = ratings_matrix.T
X.head()
```

```python
#Shape of the matrix after it is transposed.
X.shape
```

```python
#Calling a variable 
X1 = X
```

```python
#Importing the required libraries.
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
#Fitting the matrix into the model

SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix.shape
```

```python
#Correlation matrix of the matrix which was fit into the model
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```

```python
#Checking the 99th row

X.index[99]
```

```python
#It gives the number in the index when an entry is given.

i = ('970407998',
 'https://www.amazon.in/Amazin-Homes-Stylish-Printed-Computer/dp/B09ZPLLJRP/ref=sr_1_9422?crid=2A2RRWHXQP4AH&keywords=electronics&qid=1665159314&qu=eyJxc2MiOiI4LjQ0IiwicXNhIjoiNy45OCIsInFzcCI6IjUuODUifQ%3D%3D&s=electronics&sprefix=ele%2Celectronics%2C215&sr=1-9422')

product_names = list(X.index)
product_ID = product_names.index(i)
product_ID
```

```python
#Matrix of both the correlation matrix with the product_ID(variable given above)

correlation_product_ID = correlation_matrix[product_ID]
correlation_product_ID.shape
```

```python
#Final Recommendations

Recommend = list(X.index[correlation_product_ID > 0.90])

# Removes the item already bought by the customer
Recommend.remove(i) 

Recommend[0:9]
```

```python
#Recommended products as a Dataframe for easy view.

Final_DF = pd.DataFrame(Recommend, columns=['Recommended_product_ID','Recommended_Product_Links'])
Final_DF
```
